## Q1. Explain GET and POST methods.

#### Ans:
##### GET Method: The GET method is used to retrieve information from the server using a given URI. It’s safe, idempotent, and cacheable, meaning it doesn’t change any resources on the server and can be repeated without side effects. The data sent by GET method can be seen in the URL, so it should not be used to send sensitive information (like passwords).

##### POST Method: The POST method is used to send data to the server to create/update a resource. The data sent by POST method goes in the body of the HTTP request, so it’s not visible in the URL and there’s no size limitations. It’s neither safe nor idempotent, meaning it can change resources on the server and repeating it may have different effects. It’s commonly used for form submissions where user data needs to be added to the database.

## Q2. Why is request used in Flask?

### Ans:
The request object in Flask is used to get the data that a client sends to your application. It’s a global object that Flask provides which can be used to get information from the incoming HTTP request from the client. This can include form data, JSON data, headers, and other information. It’s essential for processing client requests and responding appropriately. For example, you might use request.form to access submitted form data, or request.args to access parameters in the query string of the URL. It’s a fundamental part of how Flask applications receive data from clients.

## Q3. Why is `redirect()` used in Flask?

### Ans:
The `redirect()` function in Flask is used to redirect the user to a different endpoint or URL. This is particularly useful in scenarios where after completing a certain task (like form submission), you want to take the user to a different page. For example, after a user logs in, you might want to redirect them to their dashboard. The redirect() function makes this possible by returning a response object that, when called, redirects the client to another location. It’s a key part of controlling the flow of web applications in Flask.

## Q4. What are templates in Flask? Why is the render_template() function used?

### Ans:
#### Templates:
in Flask are a way to generate dynamic HTML pages. They allow you to insert variables, control structures, and other Python code directly into your HTML, which gets evaluated before being sent to the client. This allows you to create more complex and interactive web pages.

#### The render_template():
function is used to generate these templates and send them to the client. You pass the name of the template and the variables you want to insert into the template as arguments to render_template(), and it returns a string of the generated HTML. This function is a key part of creating dynamic web applications with Flask. It allows you to keep your Python logic and HTML separate, leading to cleaner and more maintainable code. It’s one of the reasons why Flask is such a powerful tool for web development.
